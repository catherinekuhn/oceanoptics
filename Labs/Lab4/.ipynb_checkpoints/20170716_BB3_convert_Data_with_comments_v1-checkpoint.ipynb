```python
### Import the required python libraries

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
#plt.rcParams['text.usetex'] = True
#plt.rcParams['text.latex.unicode'] = True

### Import Nil's code for seawater correction
import bgc_proc as bgc

% matplotlib inline
```

```python
### Load in BB3 files using csv_read from pandas

## Read in two files for filtered and unfiltered water samples
## These two files will be compared to one another
## Modify the filepaths to reference your own files
bb3_read_unfiltered = pd.read_csv('dat/Python_blue_unfiltered_DRE_215', delimiter='\t', header=None)
bb3_read_filtered = pd.read_csv('dat/Python_blue_filtered_DRE_216', delimiter='\t', header=None)

## Read in the dark reference files
dark_correction = pd.read_csv('dat/Dark_reference', delimiter='\t', header=None)
```

```python
### Assign column names to the data frames

bb3_read_unfiltered.columns = ['filename', 'random_number', 'blue_1', 'blue_2', 'blue_3', 'blue_check', 
                             'green_1', 'green_2', 'green_3', 'green_check', 
                             'red_1', 'red_2', 'red_3', 'red_check', 'temp','other'] 

bb3_read_filtered.columns = ['filename', 'random_number', 'blue_1', 'blue_2', 'blue_3', 'blue_check', 
                             'green_1', 'green_2', 'green_3', 'green_check', 
                             'red_1', 'red_2', 'red_3', 'red_check', 'temp','other'] 

dark_correction.columns = ['filename', 'random_number', 'blue_1', 'blue_2', 'blue_3', 'blue_check', 
                             'green_1', 'green_2', 'green_3', 'green_check', 
                             'red_1', 'red_2', 'red_3', 'red_check', 'temp','other'] 
```

```python
### Step 1. Convert counts from the BB3 into VSF values (scattering attenuation)
### Beta(Theta) = (signal_count - dark) * converstion factor

## Obtain the converstion factors from instrument calibration file (current most recent coefficeints)
blue_converstion_factor = [5.13E-05, 3.59E-05, 1.89E-05]
green_converstion_factor = [0., 0., 0.]
red_converstion_factor = [5.57E-06, 4.63E-06, 3.32E-06]


## Step 1 conversion for filtered data
blue_filter_theta_104 = (bb3_read_filtered['blue_1'] - 
                                  dark_correction['blue_1']) * blue_converstion_factor[0]
blue_filter_theta_130 = (bb3_read_filtered['blue_2'] - 
                                  dark_correction['blue_2']) * blue_converstion_factor[1]
blue_filter_theta_151 = (bb3_read_filtered['blue_3'] - 
                                  dark_correction['blue_3']) * blue_converstion_factor[1]

## Step 1 conversion for unfiltered data
blue_unfiltered_theta_104 = (bb3_read_unfiltered['blue_1'] - 
                                      dark_correction['blue_1']) * blue_converstion_factor[0]
blue_unfiltered_theta_130 = (bb3_read_unfiltered['blue_2'] - 
                                      dark_correction['blue_2']) * blue_converstion_factor[1]
blue_unfiltered_theta_151 = (bb3_read_unfiltered['blue_3'] - 
                                      dark_correction['blue_3']) * blue_converstion_factor[1]                                                                                    
```

```python
### Step 2: Perform the seawater scattering correction from as described in Zhang 2009

## Reference Nil's code for obtaining the seawater scattering correction factor
## Function arguements are in order of: Temp (c), Salinity (PSU), wavelength (nm), angle, and delta term
## Leave delta term as 0.039
seawater_correction_104 = bgc.estimate_betasw(21.6, 29.72, _lambda=440, _theta=104, _delta=0.039)
seawater_correction_130 = bgc.estimate_betasw(21.6, 29.72, _lambda=440, _theta=130, _delta=0.039)
seawater_correction_151 = bgc.estimate_betasw(21.6, 29.72, _lambda=440, _theta=151, _delta=0.039)


## Subtract the salinity correction for each of the three backscatter angles
blue_filter_theta_104_sw = blue_filter_theta_104  - seawater_correction_104
blue_filter_theta_130_sw = blue_filter_theta_130 - seawater_correction_130
blue_filter_theta_151_sw = blue_filter_theta_151 - seawater_correction_151

blue_unfiltered_theta_104_sw = blue_unfiltered_theta_104 - seawater_correction_104
blue_unfiltered_theta_130_sw = blue_unfiltered_theta_130 - seawater_correction_130
blue_unfiltered_theta_151_sw = blue_unfiltered_theta_151 - seawater_correction_151
```

```python
### 3. Correct VSF for absorption along pathlength

## Beta_p(Theta) = Beta_p(Theta) e^(L( a+ epsilon * b))
## Where: L is pathlength, a is total absorption, b is total scattering, and esilon is 
## Using Wetlabs approximation of exponent


blue_filter_theta_104_sw_abs = blue_filter_theta_104_sw * np.exp(.0391*0.)
blue_filter_theta_130_sw_abs = blue_filter_theta_130_sw * np.exp(.0391*0.)
blue_filter_theta_151_sw_abs = blue_filter_theta_151_sw * np.exp(.0391*0.)

blue_unfiltered_theta_104_sw_abs = blue_unfiltered_theta_104_sw * np.exp(.0391*0.)
blue_unfiltered_theta_130_sw_abs = blue_unfiltered_theta_130_sw * np.exp(.0391*0.)
blue_unfiltered_theta_151_sw_abs = blue_unfiltered_theta_151_sw * np.exp(.0391*0.)
```

```python
### Step 4: Convert VSF at theta to bb

## Beta_p(Theta) = Beta_p(Theta) e^(L( a+ epsilon * b))
## Where: L is pathlength, a is total absorption, b is total scattering, and esilon is 
## Using Wetlabs approximation of exponent


blue_filter_theta_104_sw_abs_bbp = blue_filter_theta_104_sw_abs * 2. * np.pi * .9
blue_filter_theta_130_sw_abs_bbp = blue_filter_theta_130_sw_abs * 2. * np.pi * 1.17
blue_filter_theta_151_sw_abs_bbp = blue_filter_theta_151_sw_abs * 2.* np.pi * 1.13

blue_unfiltered_theta_104_sw_abs_bbp = blue_unfiltered_theta_104_sw_abs * 2. * np.pi * .9
blue_unfiltered_theta_130_sw_abs_bbp = blue_unfiltered_theta_130_sw_abs * 2. * np.pi * 1.17
blue_unfiltered_theta_151_sw_abs_bbp = blue_unfiltered_theta_151_sw_abs * 2.* np.pi * 1.13
```

```python
## Pplot Steps 3 and 4

plt.figure(figsize = (14,14))
plt.subplot(221)
plt.title('filtered DRE at 440 nm: VSF(Theta)')
plt.plot(blue_filter_theta_104_sw_abs, color = 'blue', label = 'Theta = 104')
plt.plot(blue_filter_theta_130_sw_abs, color = 'black', label = 'Theta = 130')
plt.plot(blue_filter_theta_151_sw_abs, color = 'red', label = 'Theta = 151')
plt.legend(loc = 'best')
#plt.xlabel(r'\textbf{time} (s)')

#plt.ylim(-.001, 0)
plt.subplot(222)
plt.title('unfiltered DRE at 440 nm: VSF(Theta)')
plt.plot(blue_unfiltered_theta_104_sw_abs, color = 'blue', label = 'Theta = 104')
plt.plot(blue_unfiltered_theta_130_sw_abs, color = 'black', label = 'Theta = 130')
plt.plot(blue_unfiltered_theta_151_sw_abs, color = 'red', label = 'Theta = 151')
plt.legend(loc = 'best')

plt.subplot(223)
plt.title('filtered DRE at 440 nm: VSF(90-180)')
plt.plot(blue_filter_theta_104_sw_abs_bbp, color = 'blue', label = 'Theta = 104')
plt.plot(blue_filter_theta_130_sw_abs_bbp, color = 'black', label = 'Theta = 130')
plt.plot(blue_filter_theta_151_sw_abs_bbp, color = 'red', label = 'Theta = 151')
plt.legend(loc = 'best')

plt.subplot(224)
plt.title('unfiltered DRE at 440 nm: VSF(90-180)')
plt.plot(blue_unfiltered_theta_104_sw_abs_bbp, color = 'blue', label = 'Theta = 104')
plt.plot(blue_unfiltered_theta_130_sw_abs_bbp, color = 'black', label = 'Theta = 130')
plt.plot(blue_unfiltered_theta_151_sw_abs_bbp, color = 'red', label = 'Theta = 151')
plt.legend(loc = 'best')

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 14}

plt.rc('font', **font)
#plt.rc('text', usetex=True)

#plt.savefig('C:/Users/Brian/Desktop/plot_1.png')
```
